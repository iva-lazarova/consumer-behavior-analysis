# Payments

## What is the preferred payment method across the states?

```python
# Most customers in the US use paypal and cradit cards
preferred_payment = data['Payment Method'].value_counts()
print(preferred_payment)
```

## Analysis of preferred payment methods across states

```python
# Group data by Location and Payment Method and count occurences
payment_methods = data.groupby(['Location', 'Payment Method']).size()
print(payment_methods)

# Reset dataframe index for readability, add Count
payment_methods = payment_methods.reset_index(name='Count')
print(payment_methods)

# Pivot the dataframe with Location as rows and Payment methods as columns 
payment_method_pivot = payment_methods.pivot(index='Location', columns='Payment Method', values='Count')
print(payment_method_pivot)
```

## Which states are the top 5 digital laggards? (cash, bank transfer)

```python
# Sum cash and bank transfer payments in a new column added to the dataframe
payment_method_pivot['Cash_and_BankTransfer'] = payment_method_pivot['Cash'] + payment_method_pivot['Bank Transfer']

# Sort the DataFrame by the combined payments column in descending order
payment_method_pivot_sorted = payment_method_pivot.sort_values(by='Cash_and_BankTransfer', ascending=False)

# Select the top 5 rows
top_5_digital_laggards= payment_method_pivot_sorted.head(5)

print("Top 5 States for Cash and Bank Transfer Payments:")
print(top_5_digital_laggards)
```

## Which are the top 5 most digitally advanced states? (mobile, online palyment)

```python
# Sum Paypal and Venmo in a new Column an add to dataframe
# Venmo is a mobile payments platgorm in the U.S.
payment_method_pivot['PayPal_and_Venmo'] = payment_method_pivot['PayPal'] + payment_method_pivot['Venmo']
#print(payment_method_pivot)

# Sort values
payment_method_pivot_sorted = payment_method_pivot.sort_values(by='PayPal_and_Venmo', ascending = False)
#print(payment_method_pivot_sorted)

# Select the top 5 rows
top_5_digital_leaders = payment_method_pivot_sorted.head(5)
print(top_5_digital_leaders)
```
